## 1. Introdução à Classificação

```python
# definir os itens
# features [1 sim, 0 não]
# pelo longo?
# perna curta?
# late?

porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]

# 1 -> porco, 0 -> cachorro
dados = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
classes = [1, 1, 1, 0, 0, 0]
```

```python
# estimador
```

```python
# usando um módulo
# pip install scikit-learn
from sklearn.svm import LinearSVC

modelo = LinearSVC()
modelo.fit(dados, classes) # treina baseado nessas classes
```

```python
animal_misterioso = [0, 1, 1]
modelo.predict([animal_misterioso]) # resultado esperado => cachorro
```

```python
animal_misterioso = [1, 1, 1]
modelo.predict([animal_misterioso]) # resultado esperado => cachorro
```

```python
animal_misterioso = [0, 0, 0]
modelo.predict([animal_misterioso]) # resultado esperado => porco
```

```python
misterio1 = [1, 1, 1]
misterio2 = [1, 1, 0]
misterio3 = [0, 1, 1]

testes = [misterio1, misterio2, misterio3]
previsoes = modelo.predict(testes)
```

```python
previsoes
```

```python
testes_classes = [0, 1, 1] # cachorro, porco, porco

# Acurácia manual
corretos = (previsoes == testes_classes).sum()
total = len(testes)
taxa_de_acerto = corretos / total * 100
taxa_de_acerto

print(f'Acurácia: {taxa_de_acerto:.2f}%')
```

```python
# Acurácia com lib
from sklearn.metrics import accuracy_score
```

```python
taxa_de_acerto = accuracy_score(testes_classes, previsoes) * 100
print(f'Acurácia: {taxa_de_acerto:.2f}%')
```

```python
# Padroẽs importantes de nome de vars

from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

# features [1 sim, 0 não]
# pelo longo?
# perna curta?
# late?

porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]

# FASE DE TREINO
treino_x = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]

# 1 -> porco, 0 -> cachorro
treino_y = [1, 1, 1, 0, 0, 0]

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)



# FASE DE TESTES
misterio1 = [1, 1, 1]
misterio2 = [1, 1, 0]
misterio3 = [0, 1, 1]

teste_x = [misterio1, misterio2, misterio3]
teste_y = [0, 1, 1] # cachorro, porco, porco
previsoes = modelo.predict(teste_x)


# FASE MÉTRICAS

# Acurácia manual
corretos = (previsoes == teste_y).sum()
total = len(teste_x)
taxa_de_acerto = corretos / total * 100
print(f'Acurácia Manual: {taxa_de_acerto:.2f}%')

# Acurácia lib
taxa_de_acerto = accuracy_score(teste_y, previsoes) * 100
print(f'Acurácia Lib: {taxa_de_acerto:.2f}%')
```
